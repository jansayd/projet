Trabalhamos em uma importadora e o preço dos nossos produtos é vinculado a cotação de:
- Dólar
- Euro
- Ouro

Para isso, vamos criar uma automação web:

- Usaremos o selenium
- Importante: baixar o webdriver

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# para rodar o chrome em 2º plano
# from selenium.webdriver.chrome.options import Options
# chrome_options = Options()
# chrome_options.headless = True # also works
# nav = webdriver.Chrome(options=chrome_options)

nav = webdriver.Chrome()

# pesquisar cotação dolar
nav.get("https://www.google.com/")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação dólar")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
cotacao_dolar = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_dolar)

# pesquisar cotação dolar
nav.get("https://www.google.com/")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação euro")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
cotacao_euro = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_euro)

# pegar cotacao ouro
site = "https://www.melhorcambio.com/ouro-hoje"
nav.get(site)
cotacao_ouro = nav.find_element_by_xpath('//*[@id="comercial"]').get_attribute("value")
cotacao_ouro = cotacao_ouro.replace(",", ".")
print(cotacao_ouro)

nav.quit()
```

### Agora vamos atualiza a nossa base de preços com as novas cotações

- Importando a base de dados

```python
import pandas as pd

tabela_produtos = pd.read_excel("produtos.xlsx")
display(tabela_produtos)
```

- Atualizando os preços e o cálculo do Preço Final

```python
tabela_produtos.loc[tabela_produtos['Moeda']=="Dólar", "Cotação"] = float(cotacao_dolar)
tabela_produtos.loc[tabela_produtos['Moeda']=="Euro", "Cotação"] = float(cotacao_euro)
tabela_produtos.loc[tabela_produtos['Moeda']=="Ouro", "Cotação"] = float(cotacao_ouro)

tabela_produtos ['Preço Base Reais'] = tabela_produtos ['Cotação'] * tabela_produtos ['Preço Base Original']
tabela_produtos ['Preço Final'] = tabela_produtos ['Ajuste'] * tabela_produtos ['Preço Base Reais']

display(tabela_produtos)
```

### Agora vamos exportar a nova base de preços atualizada

```python
tabela_produtos.to_excel("Produtos Atualizado_teste.xlsx", index=False)
```
